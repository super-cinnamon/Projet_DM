```python
df.EnvironmentSatisfaction.fillna(df.EnvironmentSatisfaction.mode, inplace=True)
df.MonthlyIncome.fillna(df.MonthlyIncome.mean, inplace = True)
df.isnull().sum()
```

```python
df1=df.select_dtypes(exclude=['object'])
fig, axs = plt.subplots(5,5, figsize= (20,20))
i = j = 0
for column in df1:
    axs[i][j].boxplot(df[column])
    i = (i+1)%5
    if i % 4 == 0:
        j = (j+1)%5
```
